```python
# deploy_cnn.py
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.applications import MobileNetV2
from tensorflow.keras import layers, models

# 1. PrГ©traitement et augmentation des donnГ©es
datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_gen = datagen.flow_from_directory(
    "fruits_dataset/",
    target_size=(128,128),
    batch_size=16,
    class_mode="categorical",
    subset="training"
)

val_gen = datagen.flow_from_directory(
    "fruits_dataset/",
    target_size=(128,128),
    batch_size=16,
    class_mode="categorical",
    subset="validation"
)

# 2. Transfer learning avec MobileNetV2
base_model = MobileNetV2(weights="imagenet", include_top=False, input_shape=(128,128,3))
base_model.trainable = False  # geler les couches du modГЁle de base

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation="relu"),
    layers.Dropout(0.5),
    layers.Dense(train_gen.num_classes, activation="softmax")
])

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

# 3. EntraГ®nement
history = model.fit(train_gen, validation_data=val_gen, epochs=30)

# 4. Sauvegarde du modГЁle entraГ®nГ©
model.save("model_cnn.h5")

print("вњ… ModГЁle CNN sauvegardГ© avec succГЁs sous 'model_cnn.h5'")
```
